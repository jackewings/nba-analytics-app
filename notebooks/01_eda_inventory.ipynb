```python
# Importing necessary libraries
import pandas as pd
```

```python
# Setting display options for pandas
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# Loading the datasets
team_stats = pd.read_csv("../data/raw/team_stats.csv")
team_stats_advanced = pd.read_csv("../data/raw/team_stats_advanced.csv")
```

```python
# Displaying the first few rows of the team_stats dataset
team_stats.head()
```

```python
# Displaying the first few rows of the team_stats_advanced dataset
team_stats_advanced.head()
```

```python
# Checking the shapes of the datasets
print("team_stats shape:", team_stats.shape)
print("team_stats_advanced shape:", team_stats_advanced.shape)
```

```python
# Checking for duplicates based on gameId and teamId
dupes_team_stats = team_stats.duplicated(subset=["gameId", "teamId"]).sum()
print("team_stats duplicate (gameId, teamId):", dupes_team_stats)

dupes_team_stats_adv = team_stats_advanced.duplicated(subset=["gameId", "teamId"]).sum()
print("team_stats_advanced duplicate (gameId, teamId):", dupes_team_stats_adv)
```

```python
# Checking for missing values in team_stats
print(team_stats.isnull().mean().round(2).sort_values(ascending=False).head(20))

# Observing the data types of the columns in team_stats (view in text editor)
team_stats.info()
```

```python
# Checking for missing values in team_stats_advanced
print(
    team_stats_advanced.isnull().mean().round(2).sort_values(ascending=False).head(20)
)

# Observing the data types of the columns in team_stats_advanced (view in text editor)
team_stats_advanced.info()
```

```python
# Converting gameDateTimeEst to datetime and checking the date range in team_stats
team_stats["gameDateTimeEst"] = pd.to_datetime(
    team_stats["gameDateTimeEst"], errors="coerce"
)
print(
    "team_stats date range:",
    team_stats["gameDateTimeEst"].min(),
    "to",
    team_stats["gameDateTimeEst"].max(),
)

# Converting gameDateTimeEst to datetime and checking the date range in team_stats_advanced
team_stats_advanced["gameDateTimeEst"] = pd.to_datetime(
    team_stats_advanced["gameDateTimeEst"], errors="coerce"
)
print(
    "team_stats_advanced date range:",
    team_stats_advanced["gameDateTimeEst"].min(),
    "to",
    team_stats_advanced["gameDateTimeEst"].max(),
)
```

```python
# Checking the number of unique games in team_stats
game_counts = team_stats["gameId"].value_counts()
print("Games with != 2 rows in team_stats:", (game_counts != 2).sum())

# Checking the number of unique games in team_stats_advanced
game_counts = team_stats_advanced["gameId"].value_counts()
print("Games with != 2 rows in team_stats_advanced:", (game_counts != 2).sum())
```

```python
# Checking the updated number of unique games in team_stats
game_counts = team_stats["gameId"].value_counts()
print("Games with != 2 rows in team_stats:", (game_counts != 2).sum())

# Checking the updated number of unique games in team_stats_advanced
game_counts = team_stats_advanced["gameId"].value_counts()
print("Games with != 2 rows in team_stats_advanced:", (game_counts != 2).sum())

# Need to filter out games with != 2 rows before proceeding with EDA and modeling
```
