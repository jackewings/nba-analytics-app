```python
# Importing necessary libraries
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting display options for pandas
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# Loading the datasets
team_stats = pd.read_csv("../data/raw/team_stats.csv")
team_stats_advanced = pd.read_csv("../data/raw/team_stats_advanced.csv")
```

```python
team_stats.info()
```

```python
# Function to convert CamelCase to snake_case
def to_snake(s):
    return re.sub(r"(?<!^)(?=[A-Z])", "_", s).lower()


# Renaming columns to snake_case for consistency
team_stats.columns = [to_snake(col) for col in team_stats.columns]
team_stats_advanced.columns = [to_snake(col) for col in team_stats_advanced.columns]
```

```python
# Function to determine the season based on the game date
def get_season(dt):
    if pd.isnull(dt):
        return None
    year = dt.year
    month = dt.month
    if month >= 10:  # October, November, December
        return f"{year}-{year + 1}"
    else:  # January–September
        return f"{year - 1}-{year}"


# Adding a season column to both datasets
team_stats["season"] = team_stats["game_date_time_est"].apply(get_season)
team_stats_advanced["season"] = team_stats_advanced["game_date_time_est"].apply(
    get_season
)
```
